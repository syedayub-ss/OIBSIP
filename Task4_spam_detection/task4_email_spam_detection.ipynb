```python
df = df[['v1', 'v2']]  # here we are taking the only two column which is required for our analysis
df.columns = ['Category', 'Message'] # assigning the coulumn names by there type
df
```

```python
df.groupby('Category').describe() # it will shows the number of values of ham and spam
```

```python
encoder = LabelEncoder() # it is used to normalize labels.

df['Category'] = encoder.fit_transform(df['Category'])

df
```

```python
from sklearn.model_selection import train_test_split #here we are training our to model by giving data to read by algorithm

X_train, X_test, y_train, y_test = train_test_split(df.Message, df.Category, test_size=0.2)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
vec = CountVectorizer()
X_train_count = vec.fit_transform(X_train.values)
X_train_count.toarray()
```

This code uses scikit-learn's CountVectorizer to convert a collection of text documents (X_train) into a matrix of token counts (X_train_count). The resulting matrix represents the frequency of each word in each document.







```python
from sklearn.naive_bayes import MultinomialNB
model = MultinomialNB()
model.fit(X_train_count, y_train)
```

```python
x_test_count = vec.transform(X_test) # to verify the model
accuracy = model.score(x_test_count, y_test)
print (f'Accuracy of the model : {accuracy*100:.2f} %')
```

```python
def predict(): #to fatch and compare the data to check weather it is a  spam are not
  message = input('Enter the message to predict : ')
  message = [message]
  vector = vec.transform(message)
  if (model.predict(vector)==1):
    print("SPAM EMAIL")
  else: print("NOT SPAM")
```
